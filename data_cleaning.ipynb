```python
import pandas as pd
import numpy as np
```

```python
file_path = 'recipes.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
quarter_size = len(df) // 4
```

```python
df_q2 = pd.read_csv(
    file_path, 
    skiprows=range(1, quarter_size + 1), 
    nrows=quarter_size
)
print(f"Loaded {len(df_q2)} recipes from the second quarter.")
```

```python
row_nan_counts = df_q2.isnull().sum(axis=1)
top_missing = df_q2.isnull().sum(axis=1).sort_values(ascending=False)
print(top_missing.head(10))
```

```python
threshold_value = int(0.7 * len(df_q2.columns))
df_balanced = df_q2.dropna(thresh=threshold_value)

print(f"Required {threshold_value} non-null values per row.")
```

```python
df_balanced.head()
```

```python
df_dropped = df.drop(columns=['AuthorId', 'AuthorName','DatePublished', 'Images', 'AggregatedRating', 'ReviewCount'])
df_dropped.head()
```

```python
unique_count = df_dropped['RecipeCategory'].nunique()
print(f"There are {unique_count} unique recipe categories.")
```

```python
top_categories = df_balanced['RecipeCategory'].value_counts().head(20).index

diverse_sample = []

for category in top_categories:
    category_slice = df_balanced[df_balanced['RecipeCategory'] == category]
    
    n_to_sample = min(5, len(category_slice))
    
    sample_chunk = category_slice.sample(n=n_to_sample)
    diverse_sample.append(sample_chunk)

df_100 = pd.concat(diverse_sample).reset_index(drop=True)

print(f"Created a diverse sample of {len(df_100)} recipes across {len(top_categories)} categories.")
```
