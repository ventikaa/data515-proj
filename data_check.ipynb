```python
file_path = "/Users/yang/Desktop/data_check.txt"

with open(file_path, "r") as f:
    data = f.read()
```

```python
blocks = data.split("-----------------------------------")
```

```python
import re
import pandas as pd

rows = []

for block in blocks:
    
    recipe_index = re.search(r"Recipe Index: (\d+)", block)
    recipe_name = re.search(r"Recipe Name: (.+)", block)
    ingredient = re.search(r"Ingredient: (.+)", block)
    products_found = re.search(r"Products Found: (\d+)", block)

    rows.append({
        "recipe_index": recipe_index.group(1) if recipe_index else None,
        "recipe_name": recipe_name.group(1) if recipe_name else None,
        "ingredient": ingredient.group(1) if ingredient else None,
        "products_found": int(products_found.group(1)) if products_found else None
    })

df = pd.DataFrame(rows)

df.head()
```

```python
df["ingredient_normalized"] = df["ingredient"]
```

```python
ingredient_map_zero = {
    "fresh cilantro stems": "cilantro",
    "Quick & Easy Tahini Sauce": "tahini sauce",
    "sultana": "golden raisins",
    "black cardamom pods": "black cardamom",
    "red capsicum": "red bell pepper",
    "Jump up and Smack Your Mom Hot Sauce": "hot sauce"
}

df["ingredient_normalized"] = df["ingredient_normalized"].replace(ingredient_map_zero)
```

```python
df[df["products_found"] == 0][["recipe_name","ingredient","ingredient_normalized"]]
```

```python
df.to_csv("/Users/yang/Desktop/clean_recipe_data.csv", index=False)
```
